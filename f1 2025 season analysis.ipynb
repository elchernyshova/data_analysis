<h1><b>F1 2025 season 

<h3><b>Welcome to the F1 2025 season analysis. Here we are going to look at some data to answer THE question: who was the best? 

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import openpyxl
```

```python
#First, let's have a look at some general data and count the points scored by each driver and team.  
season25=pd.read_excel("gps and points.xlsx")
season_25=season25.set_index('GPs')
season_25.fillna('-')
```

```python
#Now, let's make some visualizations (even though those of us fans probably know this data by heart).
plt.style.use('dark_background')
s25_top10=season_25.loc[:,'Max Verstappen':'Pierre Gasly'].sum().sort_values().plot.barh(
    figsize=(10,8),
    title='Drivers who scored points in 2025',
     color=['#52D752','#FF87BB','#006F62','#0600F0','#6692D1','#FFFFFF','#FFFFFF','#52D752','#6692D1','#006F62','#005AFF','#005AFF',
            '#00C8AA','#FF0000','#FF0000','#00C8AA','#FF8205','#0600F0','#FF8205'],
    grid=True
);
```

```python
constructors=pd.read_excel('constructors.xlsx')
constructors.set_index('position').plot.bar(
    figsize=(10,5),
    x='team',
    y='points',
    title='Constructors Championship',
    xlabel='',
    ylabel='',
    color=['#FF8205','#00C8AA','#0600F0','#FF0000','#005AFF','#6692D1','#006F62','#FFFFFF','#52D752','#FF87BB'],
    grid=True,
    legend=False
); 
```

<h3><b> As we can see, McLaren is the best team in 2025 (moral character aside). Can the same be said about the Championship Leader? Let's dive in and find out.

```python
#Let's slice some data and see the highest and lowest for our 3 championship contenders.
s25_leaders=season_25.loc[:,'Max Verstappen':'Lando Norris']
s25_leaders.style.background_gradient(cmap='RdYlGn')
```

```python
s25_leaders.plot(
    figsize=(10,5),
    title='Points scored by leaders',
    xlabel='',
    color=['#0600F0','#FF8205','Yellow'],
    style=['--'],
    grid=False
).legend(bbox_to_anchor=(1.28,1));
```

<h3><b>So, as it is seen, we can barely call it a "Lando Norris season" (or the start of "Lando Norris Era" as his biggest fan once said).
Instead of taking points as the main indicator of a worthy champion, we'll take number of wins. In the first half of the season (i.e. Australia-Great Britain) Oscar Piastri had more wins (5); in the second half (i.e. Belgium-Abu Dhabi) - Max Verstappen with 6 wins.

<h4><b>As we know: having most wins doesn't mean you're the champion (unfortunately), so let's see other indicators: fastest laps, pole positions and total number of wins in 2025.

```python
#To do that we need a new data file. 
summary25=pd.read_excel("gps wins poles laps.xlsx")
summary_25=summary25.set_index('Round')
summary_25['Fastest Lap'].value_counts()
```

```python
summary_25['Fastest Lap'].value_counts(ascending=True).plot.barh(
    figsize=(8,5),
    title='Fastest Laps',
    ylabel='',
    color=['#FF0000','#005AFF','#FF0000','#00C8AA','#0600F0','#00C8AA','#FF8205','#FF8205'],
    grid=False
);
```

<h3><b>Lando Norris and Oscar Piastri have the same number of fastest laps (6), which is understandable, as McLaren had the best car in 2025. Are fastest laps an indicator of a worthy champion? We'll leave this question without an answer.

```python
#Pole positions have to be crucial, right? Unless you're Charles Leclerc (sympathizing, not laughing). Let's count the pole-sitters in 2025.
summary_25['Pole Position'].value_counts()
```

```python
summary_25['Pole Position'].value_counts(ascending=True).plot.barh(
    figsize=(8,5),
    title='Pole Positions',
    ylabel='',
    color=['#FF0000','#00C8AA','#FF8205','#FF8205','#0600F0'],
    grid=False
);
```

<h3><b>Max Verstappen has most pole positions in 2025 - 8, while Lando Norris and Oscar Piastri in the championship winning car - 7 and 6 (respectively).

```python
#We've already looked into the number of wins in the first and second halves of the 2025 season, now it's time to do some basic math: add them.
summary_25['Winning Driver'].value_counts()
```

```python
summary_25['Winning Driver'].value_counts(ascending=True).plot.barh(
    figsize=(8,5),
    title='Winning Drivers',
    ylabel='',
    color=['#00C8AA','#FF8205','#FF8205','#0600F0'],
    grid=False
);
```

<h3><b>Lando Norris winning as many races (7, to be precise) as his teammate Oscar Piastri in the championship winning car indicates that Norris can be considered as a worthy world champion until we see Max Verstappen's number of wins (8) in his not-so-championship-winning car. 

```python
#Having covered the wins, let's find out what was the average starting position for each of our 3 championship contenders.
start_p=pd.read_excel("starting positions.xlsx")
start_leaders=start_p.loc[:,'GPs':'Lando Norris'].set_index('GPs')
start_leaders.style.background_gradient()
```

```python
avg_start_MV=start_p['Max Verstappen'].mean().round(2)
avg_start_MV
```

```python
avg_start_OP=start_p['Oscar Piastri'].mean().round(2)
avg_start_OP
```

```python
avg_start_LN=start_p['Lando Norris'].mean().round(2)
avg_start_LN
```

```python
start_leaders=start_p.loc[:,'GPs':'Lando Norris'].set_index('GPs').mean().round(2).plot.barh(
     figsize=(8,5),
    title='Average Start Position',
    ylabel='',
    color=['#0600F0','#FF8205','#FF8205'],
    grid=False
);
```

<h3><b>Max Verstappen has the worst average starting position - 3.5 - among the leaders of the 2025 season. Which works in his favor, considering that Verstappen has more wins in 2025 than anyone else. The best average starting position belongs to Oscar Piastri (2.87), which is surprising, considering the steep decline of his performance in the major part of the second half of the season. However, it shows that Piastri's first half of the season can be regarded as the strongest performance among the contenders. The average starting poisition for Lando Norris is 2.97.

```python
#Now let's move to the most controversial (or not obligatory at least) part of an f1 season - the sprints. To do that we'll need a new data file.
sprints=pd.read_excel('sprints.xlsx')
sprints.set_index('Round')
```

```python
sprints['Pole Position'].value_counts(ascending=True).plot.pie(
    title='Sprint Pole Positions',
    ylabel='',
    startangle=90,
    autopct='%.0f%%',
    colors=['#FF0000','#00C8AA','#0600F0','Orange','#FF8205']
);
```

```python
sprints['Winning Driver'].value_counts(ascending=True).plot.pie(
    title='Sprint Winners',
    ylabel='',
    startangle=90,
    autopct='%.0f%%',
    colors=['#FF0000','#FF8205','Orange','#0600F0']
);
```

<h3><b>Here are the results of the sprints: 2 out of 6 sprint pole positions with 1 sprint win for Oscar Piastri; and 2 sprint wins for each, Max Verstappen and Lando Norris.
